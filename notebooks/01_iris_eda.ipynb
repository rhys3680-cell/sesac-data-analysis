```python
from sklearn.datasets import load_iris
import pandas as pd

# DataFrame으로 바로 받기
iris = load_iris(as_frame=True)
df = iris.frame 
X = iris.data 
y = iris.target

print(iris.target_names)
print(df.head())
```

```python
df.info()
df.describe()
df['target'].value_counts()
```

```python
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df.head()
```

```python
df.groupby('species').mean(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.drop(columns='target').hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df.drop(columns='target'), hue='species')
plt.show()
```

```python
corr = df.drop(columns=['target', 'species']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='species', y='petal length (cm)')
plt.show()
```
